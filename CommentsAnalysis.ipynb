```python
#IMPORT
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.request('GET','https://www.imdb.com/title/tt5433138/reviews?ref_=tt_urv')
soup = BeautifulSoup(response.content,'html.parser')
```

```python
table = soup.find('div', attrs = {'class':'lister-list'})
review_list = []

for i in table.findAll('div', attrs = {'class':'text show-more__control'}):
  print(i.contents)
  review_list.append(i)

```

```python
print(type(review_list[0]))
```

```python
temp_list = []

for x in review_list:
  x= str(x)
  x = x.replace("<div class=\"text show-more__control\">","")
  x = x.replace("<br/>","")
  x = x.replace("</div>","")
  temp_list.append(x)

table_avis = temp_list
```

```python
table_avis[0]
```

```python
data = pd.DataFrame(table_avis,columns=['Commentaire'])
pd.set_option('display.max_colwidth', None)
data
```

```python
data[0:3]
```

```python
import re
temp = []

for txt in data["Commentaire"]:
    temp.append(re.sub("[^a-zA-Z \n]",r"",txt))

data["Commentaire"] = temp
```

```python
data["Commentaire"] = data["Commentaire"].apply(lambda x: str(x).lower())
data.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

temp_list = []
sia = SentimentIntensityAnalyzer()

def sentiment_scores(sentence):

    sentiment_dict = sia.polarity_scores(sentence)

    print("Le dictionnaire global des sentiments est : ", sentiment_dict)
    print("a t qualifie de ", sentiment_dict['neg']*100, "% Negative")
    print("a t qualifie de ", sentiment_dict['neu']*100, "% Neutre")
    print("a t qualifie de ", sentiment_dict['pos']*100, "% Positive")
    print("Phrase Globale Classe comme", end = " ")

    # decide sentiment as positive, negative and neutral
    if sentiment_dict['compound'] >= 0.05 :
        print("Positive")
        temp_list.append("Positive")

    elif sentiment_dict['compound'] <= - 0.05 :
        print("Negative")
        temp_list.append("Negative")

    else :
        print("Neutral")
        temp_list.append("Neutral")

    print("__________________________ \n")
```

```python
data["Commentaire"].apply(sentiment_scores)
```

```python
data['sentiment'] = temp_list
```

```python
#Visualisation

sns.histplot(data=data, x='sentiment', bins=range(2, 8), multiple='stack')
plt.legend(title='Sentiments')
plt.title('Histogramme des sentiments')
plt.xlabel('Valeurs')
plt.ylabel('Frquence')
plt.show()
```
